```sqlite
-- cover-page-part-1
SELECT
  e.id, d.filename, e.page,
  e.data ->> 'statement_covers_period_from' AS period_from,
  e.data ->> 'statement_covers_period_through' AS period_through,
  e.data ->> 'date_of_election' AS date_of_election,
  e.data ->> 'type_of_recipient_committee' AS committee_type,
  e.data ->> 'type_of_statement' AS statement_type,
  e.data ->> 'committee_name' AS committee_name,
  e.data ->> 'id_number' AS id_number,
  e.data ->> 'committee_street_address' AS street_address,
  e.data ->> 'committee_city' AS city,
  e.data ->> 'committee_state' AS state,
  e.data ->> 'committee_zip' AS zip,
  e.data ->> 'committee_phone' AS phone,
  e.data ->> 'committee_email' AS email,
  e.data ->> 'treasurer_name' AS treasurer_name,
  e.data ->> 'treasurer_mailing_address' AS treasurer_address,
  e.data ->> 'treasurer_city' AS treasurer_city,
  e.data ->> 'treasurer_state' AS treasurer_state,
  e.data ->> 'treasurer_zip' AS treasurer_zip,
  e.data ->> 'treasurer_phone' AS treasurer_phone,
  e.data ->> 'assistant_treasurer_name' AS assistant_treasurer,
  e.data ->> 'page_number' AS page_number,
  e.data ->> 'total_pages' AS total_pages
FROM extractions e
JOIN documents d ON e.document_id = d.id
WHERE e.classifier_data ->> 'page_type' = 'cover-page-part-1';
```
